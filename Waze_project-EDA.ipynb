```python
# Box plot
plt.figure(figsize=(5,2))
sns.boxplot(data=df, x='sessions', fliersize=3)
plt.title('Sessions Box Plot')
```

```python
# Helper function to plot the distributions
def draw_histogram(column, median_text=True, **kwargs):
    '''
    Helper function to plot the data distributions
    Arguments:
        column      - column to plot the distribution
        median_text - Toggle display of the median text
        **kwargs    - any key value arguments for the histplot
    '''
    plt.figure(figsize=(6,4))
    median = round(df[column].median(),1)                    # calculate the median
    ax=sns.histplot(data=df, x=column, **kwargs)             # plot histogram
    if median_text==True:
        plt.axvline(median, color='red', linestyle='--')     # draw the median line
        ax.text(0.20, 0.85, f'median={median}', color='red', # Add median text unless set to False
            ha='left', va='top', transform=ax.transAxes)
    else:
        print('Median:', median)
    plt.title(f'{column} Distribution Using Histogram', fontsize=14)
```

### Handling Outliers

```python
# Create a helper function to replace outliers with 95th percentile
def replace_outliers(column):
    '''
    Helper function to replace outliers with 95th percentile value
    Arguments:
        column - column to replace outlier
    '''
    percent_95 = df[column].quantile(0.95)
    df.loc[df[column]>percent_95, column] = percent_95
```

```python
# Replace outliers
columns = ['sessions', 'drives', 'total_sessions', 'driven_km_drives', 'duration_minutes_drives']
for column in columns:
    replace_outliers(column)
```

### Hypothesis Testing

```python
# 1. Create `map_dictionary`
map_dictionary = {'Android':1, 'iPhone':2}

# 2. Create new `device_type` column
### YOUR CODE HERE ###
df['device_type'] = df['device']

# 3. Map the new column to the dictionary
### YOUR CODE HERE ###
df['device_type'] = df['device_type'].map(map_dictionary)
```

```python
# Mean drives per device
df.groupby('device_type')['drives'].mean()
```

Based on the averages shown, it appears that drivers who use an iPhone device to interact with the application have a higher number of drives on average. However, this difference might arise from random sampling, rather than being a true difference in the number of drives. To assess whether the difference is statistically significant, we can conduct a hypothesis test.

$H_0$ = There is no difference in average number of drives based on device type.<br>
$H_A$ = There is difference in average number of drives based on device type.

```python
# Isolate the `drives` column for iPhone users.
iPhone = df[df['device_type']==2]['drives']

# Isolate the `drives` column for Android users.
Android = df[df['device_type']==1]['drives']

# Perform the two sample t-test
stats.ttest_ind(a=iPhone, b=Android, equal_var=False)
```

The observed p-value is 0.07 i.e. 7% which is greater than the chosen 5% significance level. Therefore, we fail to reject null hypothesis and lack statistical significance that drives differ with device type.
